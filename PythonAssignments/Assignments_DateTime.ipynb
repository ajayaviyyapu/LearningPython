```python
from datetime import date,datetime

#Get the current day, month, year, hour, minute and timestamp from datetime module
p = print

datetoday = datetime.now()
p(datetoday.day)
p(datetoday.month)
p(datetoday.year)
p(datetoday.hour)
p(datetoday.minute)
p(datetoday.timestamp())

#Format the current date using this format: "%m/%d/%Y, %H:%M:%S")

print(datetoday.strftime("%m/%d/%Y, %H:%M:%S"))

#Today is 5 December, 2019. Change this time string to time.

str_date  = '5 December, 2019'

date_date = datetime.strptime(str_date,'%d %B, %Y')

print(date_date)

#Calculate the time difference between now and new year.

date_today = datetime.now()
date_new_year = datetime(year=2027, month=1, day=1)

difference = date_new_year - date_today

p(difference)

#Calculate the time difference between 1 January 1970 and now.

date_given = date(year=1970, month=1, day=1)

date_today = date.today()

difference = date_today - date_given

p(difference)
```
