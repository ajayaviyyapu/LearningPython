```python
#Write a function which count number of lines and number of words in a text. All the files are in the data the folder:

#Read obama_speech.txt file and count number of lines and words
#Read michelle_obama_speech.txt file and count number of lines and words
#Read donald_speech.txt file and count number of lines and words
#Read melina_trump_speech.txt file and count number of lines and words

import os
import numpy as np
import re
import functools as ft
import json
from collections import defaultdict

with open('/home/TestData/obama_speech.txt') as f :
  lines = f.readlines()
  print(f'line count {len(lines)}')

print(lines)

def removeSpecialCharacters(line) :
  count_word = 0
  match_exp = r'[^A-Za-z0-9 ]*'
  special_chracters = re.findall(match_exp,line)
  for spc in special_chracters :
    line = line.replace(spc,'')
    list_words = line.split(" ")
  for word in list_words :
    if word and word is not None :
      count_word += 1
  return count_word

words_list = ft.reduce(lambda x,y : x+y,list(map(removeSpecialCharacters,lines)))

print(words_list)

#Read the countries_data.json data file in data directory, create a function that finds the ten most spoken languages

with open('/home/TestData/countries_data.json') as f :
  country_data = json.load(f)
print(country_data)

lang_dict = defaultdict(int)

for country in country_data :
  if country.get('languages') and country.get('languages') is not None :
    for lang in country.get('languages') :
      lang_dict[lang] += 1

top_ten_spoken_langs = sorted(lang_dict.items(), key= lambda item:item[1], reverse=True)[0:10]

print(top_ten_spoken_langs)

#Read the countries_data.json data file in data directory, create a function that creates a list of the ten most populated countries as a json

population_dict = defaultdict(int)

for country in country_data :
  if country.get('population') and \
    country.get('population') is not None and \
    country.get('name') and \
    country.get('name') is not None :
      population_dict[country.get('name')] = country.get('population')

top_ten_populous_countries = sorted(population_dict.items(), key= lambda item:item[1], reverse=True)[0:10]

print(top_ten_populous_countries)
top_ten_populous_countries_json = list(map(lambda country: {'country':country[0],'population':country[1]}, top_ten_populous_countries))

output = "[\n"

items = []
for item in top_ten_populous_countries_json:
    items.append(f"{{'country': '{item['country']}', 'population': {item['population']}}}")

output += ",\n".join(items)
output += "\n]"
print(output)

jsonstr = json.dumps(top_ten_populous_countries_json, indent=1,separators=(', ', ': '))
print(jsonstr)

with open('/home/TestData/top_pop_ctrs.json','w') as f :
  json.dump(top_ten_populous_countries_json,f,indent=2)


#Extract all incoming email addresses as a list from the email_exchange_big.txt file.

num_list = range(0,5)

numpy_Array = np.array(num_list)

print(numpy_Array ** 2)

#Extract all incoming email addresses as a list from the email_exchange_big.txt file
email_list=list()

with open('/home/TestData/email_exchanges_big.txt') as f :
    while True :
      line = f.readline()
      if line is None or not line :
        break
      if line.count('From:') > 0 :
        line_email_interim = line.replace('From: ','')
        line_email = line_email_interim.replace('\n', '')
        email_list.append(line_email)

print(email_list)
print(len(email_list))

#Find the most common words in the English language.
#Call the name of your function find_most_common_words, it will take two parameters - a string or a file and a positive integer, indicating the number of words.
#Your function will return an array of tuples in descending order.

word_count = defaultdict(int)
line_list = list()

def find_most_common_words(filename,num_words) :
  word_count.clear()
  with open('/home/TestData/' + filename) as f :
      line_list = f.readlines()
  reg_mat = r'[^A-Za-z ]'
  for line in line_list :
    special_symbols = re.findall(reg_mat,line)
    #print(special_symbols)
    for symbol in special_symbols :
      line.replace(symbol,'')
    line = line.replace('\n','')
    word_list = line.split(' ')
    #print(word_list)
    #print(word_count)
    for word in word_list :
      if word and word is not None :
        word_count[word] += 1

  word_list_num = sorted(word_count.items(),key= lambda item:item[1], reverse=True)[0:num_words]
  return word_list_num


print(find_most_common_words('sample.txt',10))

print(find_most_common_words('sample.txt',5))


```
